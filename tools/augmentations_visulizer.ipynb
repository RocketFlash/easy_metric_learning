```python
import albumentations as A
import cv2
from pathlib import Path
from matplotlib import pyplot as plt
```

```python
img_h = 224
img_w = 224

image_path = Path('/datasets/metric_learning/dataset_v0/000000044004/gmi_product_image_KltNLrM.png')
image = cv2.imread(str(image_path))
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
plt.show()
```

```python
transforms = A.Compose([
    # A.Resize(
    #     height=img_h,
    #     width=img_w,
    #     p=1
    # ),
    A.LongestMaxSize(img_h),
    A.PadIfNeeded(
        min_height=img_h, 
        min_width=img_w, 
        border_mode=cv2.BORDER_CONSTANT,
        value=(127,127,127)
    ),
    A.RandomRotate90(p=1),
    # A.HorizontalFlip(p=0.5),
    A.ShiftScaleRotate(
        rotate_limit=15, 
        scale_limit=0.1,
        shift_limit=0.05,
        border_mode=2, 
        p=0.5
    ),
    # A.RandomBrightnessContrast(
    #     brightness_limit=[-0.15, 0.15],
    #     contrast_limit=[-0.15, 0.15],
    #     p=1
    # ),
    # A.Blur(
    #     blur_limit=[3, 5],
    #     p=1
    # ),
    # A.CoarseDropout(
    #     max_holes=8,
    #     min_holes=4,
    #     max_height=40,
    #     max_width=40,
    #     min_height=20,
    #     min_width=20,
    #     p=1
    # ),
    # A.GaussNoise(
    #     var_limit=(10.0, 50.0),
    #     p=1
    # ),
    # A.ISONoise(
    #     intensity=[0.1, 0.5],
    #     color_shift=[0.01, 0.05],
    #     p=1
    # ),
    # A.MotionBlur(
    #     blur_limit=[3, 5],
    #     p=1
    # )
    # A.CLAHE(p=1),
    # A.Sharpen(p=1),
    # A.ToGray(p=1)
    # A.ColorJitter(
    #     brightness=0.1, 
    #     contrast=0.1, 
    #     saturation=0.1,
    #     p=1
    # ),
    
])


N = 5
f, axarr = plt.subplots(1, N, figsize=(12, 12))

for i in range(N):
    sample = transforms(image=image)
    image_transformed = sample['image']
    axarr[i].imshow(image_transformed)   
    
plt.show()
```
